```python
from langchain_community.llms import HuggingFacePipeline
from transformers import pipeline
```

```python
pipe = pipeline("text-generation", model="gpt2")

llm = HuggingFacePipeline(pipeline=pipe)
```

```python
print(llm.invoke("Explain Artificial Intelligence in simple terms"))
```

```python
topic = "Machine Learning"
prompt = f"Explain {topic} in simple terms"

print(prompt)
print(llm.invoke(prompt))
```

```python
# Simple chain-like function (replacement for LLMChain)

def simple_chain(topic):
    prompt = f"Explain {topic} in simple terms"
    return llm.invoke(prompt)

# Run it
result = simple_chain("Deep Learning")
print(result)
```

```python
# Simple memory using Python list

memory = []

# Store conversation
memory.append(("Hi", "Hello"))
memory.append(("What is AI?", "AI is Artificial Intelligence"))

# Print memory
for inp, out in memory:
    print(f"User: {inp}")
    print(f"Bot: {out}")
    print("---")
```

```python
# Simple chatbot simulation using memory

memory = []

def chatbot(user_input):
    response = llm.invoke(user_input)
    memory.append((user_input, response))
    return response

# Conversation
print(chatbot("Hi"))
print(chatbot("What is Machine Learning?"))

# Show conversation history
print("\nConversation History:")
for inp, out in memory:
    print(f"User: {inp}")
    print(f"Bot: {out}")
    print("------")
```
